Dùng Random Walk để tạo thành các câu từ các labels thuộc node kề nhau

```python
import os
import networkx as nx
import random
import csv
# Thư mục chứa các file GML
folder_path = '/content/graph/'

# Hàm để đọc đồ thị từ một file GML
def read_gml_file(file_path):
    return nx.read_gml(file_path)

# Hàm thực hiện Random Walk
def random_walk(graph, start_node, num_steps):
    current_node = start_node
    index = current_node.find('(')
    walk = [current_node[:index]]
    for _ in range(num_steps):
        neighbors = list(graph.neighbors(current_node))
        if neighbors:
            current_node = random.choice(neighbors)
            index = current_node.find('(')
            if current_node[:index] not in walk:
              walk.append(current_node[:index])
        else:
            break
    return walk

sentences_labels = [] # <- Này để train model word2vec


# Lặp qua tất cả các tệp trong thư mục

# Xử lý
for file_name in os.listdir(folder_path):
    if file_name.endswith('.gml'):
        file_path = os.path.join(folder_path, file_name)
        # Đọc đồ thị từ file GML
        graph = read_gml_file(file_path)
        # Thực hiện Random Walk từ mỗi nút trong đồ thị và lưu trữ kết quả
        for node in graph.nodes:
            num_steps = 12   # Số bước trong mỗi Random Walk
            walk = random_walk(graph, node, num_steps)
            sentences_labels.append(walk)

# Ghi dữ liệu vào file json
json_file_path = '/content/All_sentences.json'
with open(json_file_path, mode='w', newline='') as file:
    json.dump(sentences_labels,file)

print("Đã lưu vào file json thành công!")
```

```python
# thống kê số độ dài các câu sau khi sử dụng random walk
# qua đó lấy số liệu để train model word2vec
ls={}
for walk in sentences_labels:
  if len(walk) not in ls:
    ls[len(walk)] = 1
  else:
    ls[len(walk)] += 1
print(ls)
print(len(sentences_labels))
```

```python
# Ma tran 2 chieu -> 3 chieu
def expand_adjacency_matrix(adj_mat: torch.Tensor, n_heads: int) -> torch.Tensor:

    adj_mat = torch.tensor(adj_mat)
    # Get the number of nodes in the graph
    n_nodes = adj_mat.size(0)

    # Add a new dimension to the adjacency matrix to make it 3D
    adj_mat_expanded = torch.unsqueeze(adj_mat, dim=2)

    # Repeat the adjacency matrix along the third dimension to match the number of heads
    adj_mat_expanded = adj_mat_expanded.repeat(1, 1, n_heads)

    return adj_mat_expanded
```

```python
import torch
from torch import nn

from labml_helpers.module import Module

class GraphAttentionLayer(Module):
  def __init__(self, in_features: int, out_features: int, n_heads: int,is_concat: bool = True,dropout: float = 0.6,leaky_relu_negative_slope: float = 0.2):
    super().__init__()
    self.is_concat = is_concat
    self.n_heads = n_heads
    if is_concat:
      assert out_features % n_heads == 0
      self.n_hidden = out_features // n_heads
    else:
      self.n_hidden = out_features
    self.linear = nn.Linear(in_features, self.n_hidden * n_heads, bias=False)
    self.attn = nn.Linear(self.n_hidden * 2, 1, bias=False)
    self.activation = nn.LeakyReLU(negative_slope=leaky_relu_negative_slope)
    self.softmax = nn.Softmax(dim=1)
    self.dropout = nn.Dropout(dropout)
  def forward(self, h: torch.Tensor, adj_mat: torch.Tensor):
    n_nodes = h.shape[0]
    g = self.linear(h).view(n_nodes, self.n_heads, self.n_hidden)
    g_repeat = g.repeat(n_nodes, 1, 1)
    g_repeat_interleave = g.repeat_interleave(n_nodes, dim=0)
    g_concat = torch.cat([g_repeat_interleave, g_repeat], dim=-1)
    g_concat = g_concat.view(n_nodes, n_nodes, self.n_heads, 2 * self.n_hidden)
    e = self.activation(self.attn(g_concat))
    e = e.squeeze(-1)
    assert adj_mat.shape[0] == 1 or adj_mat.shape[0] == n_nodes
    assert adj_mat.shape[1] == 1 or adj_mat.shape[1] == n_nodes
    assert adj_mat.shape[2] == 1 or adj_mat.shape[2] == self.n_heads
    e = e.masked_fill(adj_mat == 0, float('-inf'))
    a = self.softmax(e)
    print(a)
    a = self.dropout(a)
    print(a)
    attn_res = torch.einsum('ijh,jhf->ihf', a, g)
    if self.is_concat:
      return attn_res.reshape(n_nodes, self.n_heads * self.n_hidden)
    else:
      return attn_res.mean(dim=1)
```
